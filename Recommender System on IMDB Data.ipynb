```python
import re
```

```python
def clean_date_range(date_range):
    #removing symbols
    cleaned_range = re.sub(r'[\(\)–\s]', '', date_range)
    #spliting to start and end year
    start_year = cleaned_range[:4]
    end_year = cleaned_range[4:]
    if end_year == '':
        end_year = 'Present'
    return f"{start_year}-{end_year}"
    
```

```python
cleaned_array = [clean_date_range(date_range) for date_range in df2["Runtime_of_Series"]]
```

```python
def buld_chart(df2,percentile=0.08):
    print("Preferred Genre: ")
    genre = input().capitalize()
    print("Preferred Certificate: ")
    certificate = str(input())
    print("Release Year)")
    year = str(input())
    movies=df2.copy()
    movies=movies[(movies["Genre"]==genre) & (movies["Certificate"]==certificate) & (movies["Release_year"]==year)]
    C = movies["IMDB_Rating"].mean()
    m = movies["No_of_Votes"].quantile(percentile)
    q_movies = movies.copy().loc[movies['No_of_Votes']>=m]
    q_movies['score'] = q_movies.apply(lambda x: (x['No_of_Votes']/(x['No_of_Votes']+m)* x['IMDB_Rating']) + (m/(m+x['No_of_Votes'])*C),axis=1)
    q_movies = q_movies.sort_values(by='score',ascending=False)
    return q_movies
        
    
```
